```python
import re                                 
import string                            

from nltk.corpus import stopwords         
from nltk.stem import PorterStemmer       
from nltk.tokenize import TweetTokenizer 
```

```python
soz2=''
for soz in sozder:
    soz2 += re.sub(r'^RT[\s]+', '', soz)
# remove hyperlinks
soz2 = re.sub(r'https?:\/\/.*[\r\n]*', '', soz2)

soz2 = re.sub(r'\n', '', soz2)

soz2 = re.sub(r'\d+', '', soz2)
    
soz2 = re.sub(r'<', '', soz2)
    
soz2 = re.sub(r'>', '', soz2)
    
soz2 = re.sub(r'=', '', soz2)
    
soz2 = re.sub(r'/', '', soz2)
    
print(soz2)
```

```python
tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True,
                               reduce_len=True)

# tokenize tweets
all_tokens = tokenizer.tokenize(soz2)

print()
print('Tokenized string:')
print(all_tokens)
```

```python
soz_clean = []
stopwords_english = stopwords.words('english')
for word in all_tokens: # Go through every word in your tokens list
    if (word not in stopwords_english and  # remove stopwords
        word not in string.punctuation):  # remove punctuation
        soz_clean.append(word)

print('removed stop words and punctuation:')
print(soz_clean)
```

```python


# Instantiate stemming class
stemmer = PorterStemmer() 

# Create an empty list to store the stems
all_stem = [] 

for word in soz_clean:
    stem_word = stemmer.stem(word)  # stemming word
    all_stem.append(stem_word)  # append to the list

print('stemmed words:')
print(all_stem)
```
